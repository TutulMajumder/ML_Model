#### Random forest

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
X.shape
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
df.head()
```

```python
df.shape
```

#### function for row sampling

```python
def sample_rows(df,percent):
    return df.sample(int(percent*df.shape[0]),replace=True)
```

```python
rdf1=sample_rows(df,0.2)
```

```python
rdf2=sample_rows(df,0.2)
```

```python
rdf3=sample_rows(df,0.2)
```

```python
print(rdf1.columns)
print(rdf2.columns)
print(rdf3.columns)
```

```python
rdf1
```

```python
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
clf1.fit(rdf1.iloc[:,0:5],rdf1.iloc[:,-1])
clf2.fit(rdf2.iloc[:,0:5],rdf1.iloc[:,-1])
clf3.fit(rdf3.iloc[:,0:5],rdf1.iloc[:,-1])
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```

#### Function for feature sampling

```python
df.columns.to_list()[:-1]
```

```python
df.shape[1]-1
```

```python
import random
```

```python
def sample_feature(df,percent):
    cols=random.sample(df.columns.to_list()[:-1],int(percent*(df.shape[1]-1)))
    new_df = df[cols]
    new_df['target']=df['target']
    return new_df
```

```python
fdf1=sample_feature(df,0.8)
```

```python
fdf2=sample_feature(df,0.8)
```

```python
fdf3=sample_feature(df,0.8)
```

```python
fdf1
```

```python
fdf2
```

```python
fdf3
```

```python
clf1.fit(fdf1.iloc[:,0:4],fdf1.iloc[:,-1])
clf2.fit(fdf2.iloc[:,0:4],fdf1.iloc[:,-1])
clf3.fit(fdf3.iloc[:,0:4],fdf1.iloc[:,-1])
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```

#### Combined sample

```python
def combined_sample(df,row_percent,col_percent):
    new_df = sample_rows(df,row_percent)
    return sample_feature(new_df,col_percent)
```

```python
cdf1=combined_sample(df,0.5,0.5)
```

```python
cdf2=combined_sample(df,0.5,0.5)
```

```python
cdf3=combined_sample(df,0.5,0.5)
```

```python
print(cdf1.columns)
print(cdf2.columns)
print(cdf3.columns)
```

```python

clf1.fit(cdf1.iloc[:,0:2],cdf1.iloc[:,-1])
clf2.fit(cdf2.iloc[:,0:2],cdf2.iloc[:,-1])
clf3.fit(cdf3.iloc[:,0:2],cdf3.iloc[:,-1])
```
